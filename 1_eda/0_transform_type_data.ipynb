# Transformación de los datos a tipo correcto

Este notebook contiene únicamente los datos transformados al tipo de dato correcto. En los datos raw muchos de los datos están en formato erróneo por lo que es necesario trasformarlo a formato correcto (ej de string a integer en datos de ventas) para que las proximas veces que sea lean los datos ya se haga con el formato correcto

```python
import pandas as pd
import pickle
import numpy as np
```

```python
import os
# set path root of repo
actual_path = os.path.abspath(os.getcwd())
list_root_path = actual_path.split('/')[:-1]
root_path = '/'.join(list_root_path)
os.chdir(root_path)
print('root path: ', root_path)
```

### 1. Read raw files

```python
folder_data = 'data/data_input_raw/'
df_cal = pd.read_csv(folder_data + 'calendar.csv')
df_eval = pd.read_csv(folder_data + 'sales_train_evaluation.csv')
df_price = pd.read_csv(folder_data + 'sell_prices.csv')
df_sample_output = pd.read_csv(folder_data + 'sample_submission.csv')

df_validation = pd.read_csv(folder_data + 'sales_train_validation.csv')
```

```python
# set path to raw files
input_path = 'data/data_input_raw/'
output_path = 'data/data_input_dtype/'
```

### 2. Transformar: calender

```python
# read dtypes correct
cal_dtypes = {
    'd': 'category',
    'wm_yr_wk': np.uint16,    
    'event_name_1': 'category',
    'event_type_1': 'category',
    'event_name_2': 'category',
    'event_type_2': 'category',
    'snap_CA': np.uint8,
    'snap_TX': np.uint8,
    'snap_WI': np.uint8,
}

cal = pd.read_csv(input_path + 'calendar.csv', 
                  dtype=cal_dtypes, 
                  #usecols=list(cal_dtypes.keys()) + ['date'], 
                  usecols=list(cal_dtypes.keys()) + ['date'] + ['weekday','wday','month','year'], 
                  parse_dates=['date'])

event_cols = [k for k in cal_dtypes if k.startswith('event')]

for col in event_cols:
    cal[col] = cal[col].cat.add_categories('nan').fillna('nan')
```

```python
# nuevo df
cal.info()
```

```python
# show new df
cal.head()
```

```python
# save
cal.to_pickle(output_path + 'calendar.pkl')
```

### 3. Transform Price

```python
prices_dtypes = {
    'store_id': 'category',
    'item_id': 'category',
    'wm_yr_wk': np.uint16,
    'sell_price': np.float32
}
prices = pd.read_csv(input_path + 'sell_prices.csv', dtype=prices_dtypes)
```

### 4. Transform Sales

```python
sales_dtypes = {
    'id': 'category',
    'item_id': prices.item_id.dtype,
    'dept_id': 'category',
    'cat_id': 'category',
    'store_id': 'category',
    'state_id': 'category',
    **{f'd_{i}': np.float32 for i in range(1942)}
}
sales = pd.read_csv(
    input_path + 'sales_train_evaluation.csv',
    dtype=sales_dtypes,
)
```
